<h3>Regression Line with seaborn </h3>

<a href="https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html"> scipy ref for reg params </a>

```python
w= np.array([73.84,68.17,74.11,71.73,69.88,67.53,68.78])
h= np.array([241,162,212,220,206,152,184])

import seaborn as sns
ax = sns.regplot(h,w, color='aqua', marker='^' )

#get the reg line parameters:         using linregress from scipy.stats
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(h,w)
print("slope: {:.2f}, intercept: {:.2f}".format(slope, intercept))
print("R value: {:.2f}, p value: {:.2f}".format(r_value, p_value))
print("predicted values for h")
for h,pw in zip(h, h*slope+intercept):
    print("{:.2f}  {:.2f}".format(h, pw))
```

```python
def plt_2_pt_line(L):
    ''' L is a list of two tuple lists, points of a line like [(1,2),(3,4)] '''
    fig=plt.figure()
    ax=plt.axes()
    ax.axhline(color='gray')
    ax.axvline(color='gray')
    for points in L:
        plt.plot( [points[0][0], points[1][0]  ] , [  points[0][1], points[1][1] ] )
   


```

<h2>Density and Contour Plots: 3D in 2D </h2>

```python
#a 3D function:
def f(x, y):
    return np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x) 
xy=[(1,1),(1,2),(2,1),(2,2)]
Z= [np.round(f(x,y),3) for (x,y) in  xy]
print(Z)
```

<h4> mesh the grid </h4> `meshgrid(x,y)`
builds two-dimensional grids from one-dimensional arrays:

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black')
```

negative values in dashed lines 

`cmap`



```python
nlines= 20
plt.contour(X, Y, Z, 30, cmap='RdGy')# RdGy (short for Red-Gray) colormap
#nlines equally spaced intervals within the data range
```

`colorbar()` automatically creates additional axis labeled color info

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

<h4> Combine contour plots and image plots </h4>
`plt.clabel()`

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
